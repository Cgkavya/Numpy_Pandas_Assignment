```python
# Convert ? to zero and convert horse power field to float and find its mean

power = data['horsepower'].to_numpy()
hp = np.where(power == '?', '0', power)
data['horsepower'] = hp.astype(float)
m = np.mean(data['horsepower'])
print(m)
```

```python
# Replace zero by mean
#data[data['horsepower'] == '0'] = m
horse_power = np.where(data['horsepower'] == '0', m, data['horsepower'])
```

```python
# Create new column with new values

data['new_horsepower'] = data['horsepower'] + (0.1 * data['horsepower'])
```

```python
# Find the average displacement of cars with an origin of 2
boolean_mask = data['origin'] == 2
r = data['displacement'][boolean_mask]
print(r)
avg = np.mean(r)
print(avg)
```
