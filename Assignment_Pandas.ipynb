```python
 # check for missing values in horsepower columns
missing_values = data[data["horsepower"] == '?']
missing_values
```

```python
# Replace missing values

cleaned_data = data['horsepower'].replace('?', '0')
#data[data['horsepower'] =='?'] = 0
#data['horsepower'] = data['horsepower'].astype(float)
h = cleaned_data.astype(float)
#m = data['horsepower'].median()
m = h.median()
print(m)
```

```python
# Replace zero by median
h.replace('0', m)
h.values
```

```python
# Create new column with new values
data['power_to_weight_ratio'] = h /  data['weight']
data
```

```python
# Find mean of mpg grouping origin
g = data.groupby(['origin'])['mpg'].mean()
print(g)
```

```python
# sort by mpg in desc order and get first 10 car names
sorted_df = data.sort_values(by='mpg',ascending=False)
#sorted_df.drop(sorted_df.index[0:6], inplace=True)
s = sorted_df[:10]
car_name = s.iloc[:,8]
print(car_name)
```

```python
# Create a new column performance_score using a custom function
def performance_score(row): 
 return row['mpg'] * row['acceleration'] / row['weight'] 
data['performance_score'] = data.apply(performance_score, axis = 1)
for res in data['performance_score']:
    print(res)
```

```python
# Generate a summary DataFrame with: Average mpg, weight, and horsepower for each model_year. 
data['horsepower'] = h
a = data.groupby(['model year'])['mpg'].mean()
b = data.groupby(['model year'])['weight'].mean()
c = data.groupby(['model year'])['horsepower'] .mean()
#print(a,b,c)
array = pd.DataFrame([a,b,c])
print(array)
array.info()
```

```python
# Calculate the IQR for mpg. 
mpg_q1 = data["mpg"].quantile(0.25)
mpg_q3 = data["mpg"].quantile(0.75)
IQR = mpg_q3 - mpg_q1
upper_range = mpg_q3 + 1.5 * IQR
lower_range = mpg_q1 - 1.5 * IQR
print(mpg_q1, mpg_q3, IQR, upper_range, lower_range)
```

```python
# Define outlier
mpg_outlier = data["mpg"][(data["mpg"] > upper_range) |
                                   (data["mpg"] < lower_range)] 

mpg_outlier
```
